```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

#通过api的方式

#构造请求头
url = 'https://top.baidu.com/api/board?tab=car'

header = {
    'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/116.0.0.0 Safari/537.36', 
    'Host' : 'top.baidu.com',
    'Referer': 'https://www.google.com/',
    'Accept-Encoding': 'gzip, deflate, br',
    'Accept-Language': 'zh-CN,zh;q=0.9',
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.7',
}

#发送请求
r = requests.get(url,header)

#检验返回的数据类型
#print(r.text)

#获得json数据
data_json = r.json()
print(data_json)
print(type(data_json))
```

```python
#查看字典中的key
# 使用 for 循环遍历字典的键
for key in data_json:
    print(key)
    
```

```python
#查看data的内容，使用get方法
data_value = data_json.get('data')
print(data_value)
```

```python
#查看data字典中的key
for key in data_value:
    print(key)
```

```python
# 查看cards中的内容
cards = data_value['cards']
print(cards)
```

```python
#查看content
content = cards[0]['content']
print(content)
    
```

```python
#content为一个字典，检索第一个值。
content_0 = data_value["cards"][0]['content'][0]
print(content_0)
```

```python
#爬取热搜
 
appUrl_list = []
hotChange_list = []
hotScore_list = []
index_list = []
price_list = []
show_info = []
word_list = []
level_list = []

for item in content:
    appUrl_list.append(item['appUrl'])
    hotChange_list.append(item['hotChange'])
    hotScore_list.append(item['hotScore'])
    index_list.append(item['index']+1)#排名从1开始排
    price_list.append(item['price'])
    show_info_text = item['show'][1].replace('级别：', '') # 提取级别信息并去掉"级别："
    show_info.append(show_info_text)
    word_list.append(item['word'])
    
```

```python
#拼装数据
df = pd.DataFrame(
    {
        '排名': index_list,
        '汽车名': word_list,
        '价格': price_list,
        '汽车类别'
        '热搜指数': hotScore_list,
        '汽车链接': appUrl_list,
        '热搜状态': hotChange_list,
        '汽车类型': show_info
    }
)

print(df)
# 保存为Excel文件
df.to_excel('汽车热搜榜单_api方法.xlsx', index=False)
```
