```python
 # 构造请求头、发送请求
import pandas as pd
import requests
from bs4 import BeautifulSoup

#通过api的方式

#构造请求头
url = 'https://top.baidu.com/board?tab=car'

header = {
    'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/116.0.0.0 Safari/537.36', 
    'Host' : 'top.baidu.com',
    'Referer': 'https://www.google.com/',
    'Accept-Encoding': 'gzip, deflate, br',
    'Accept-Language': 'zh-CN,zh;q=0.9',
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.7',
}

#发送请求
r = requests.get(url,header)
html_doc = r.text
print(html_doc)
```

```python
#使用soup.find_all方法，寻找某个元素组。
soup = BeautifulSoup(html_doc,'html.parser',from_encoding = 'utf-8')
div_list_1 = soup.find_all('div', class_='category-wrap_iQLoo horizontal_1eKyQ')

item_list = []
for div in div_list_1:
    #汽车名称
    name = div.find(class_='c-single-text-ellipsis').get_text()
    
    #汽车链接,查找<div>元素下，所有<a>标签。
    a_tags = div.find_all('a',href=True) #查找带有href属性的<a>标签
    for a in a_tags:
        url =a['href']  # 获取href属性内容
        
    #热搜指数
    hot_index = div.find(class_='hot-index_1Bl1a').get_text()
    
    
    item = [name, url,hot_index]
    item_list.append(item)


    #级别、价格 这段代码不知道为什么会爬取出更多的信息
    intro_divs = div.find_all('div', class_='intro_1l0wp')
    level = None
    price = None
    
    for i in intro_divs:
        text = i.get_text().strip()
        # 根据文本内容进行判断并提取信息
        if "级别：" in text:
            level = text.replace("级别：", "")
        elif "价格：" in text:
            price = text.replace("价格：", "")
        
    # print(level)
    print(name, level, price, hot_index,url)

# cols = ['汽车名称','汽车链接','热搜指数']
# df = pd.DataFrame(item_list,columns=cols)
# df.to_excel('汽车热搜榜单_html爬取.xlsx', index = False)
```
